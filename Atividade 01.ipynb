10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
engine = create_engine('sqlite:///dados/atividade1/sqlite-sakila.db')
insp = inspect(engine)
```

```python
insp.get_table_names()
```

```python
sql_query = '''SELECT * FROM country JOIN city ON country.country_id = city.country_id'''
city_country_df = pd.read_sql_query(sql_query, engine)
city_country_df.head(2)
```

```python
city_country_df.loc[city_country_df['country'].str.startswith('B'), ['city', 'country']]
```
